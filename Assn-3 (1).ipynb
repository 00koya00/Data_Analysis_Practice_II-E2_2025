### Figure 2

Use `data-fig2.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'Assn-3-data/data-fig2.npy' )

R = []
G = []
B = []
for i in range (0, len(a)):
    R.append(a[i,0])
    G.append(a[i,1])
    B.append(a[i,2])

plt.plot(R, color = 'red', label="Red" )
plt.plot(G, color = 'green', label="Green" )
plt.plot(B, color = 'blue', label="Blue" )

plt.xlabel("X value")
plt.ylabel("Y value")
plt.legend(loc=2)
plt.show()
```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'Assn-3-data/data-fig3.npy' )

x = []
y = []

for i in range (0, len(a)):
    x.append(a[i,0])
    y.append(a[i,1])

plt.scatter(x,y, s = 150, c = "blue", edgecolors = "blue")

plt.xlabel("X value")
plt.ylabel("Y value")
plt.show()
```
